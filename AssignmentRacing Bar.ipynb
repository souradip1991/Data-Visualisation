```python
import pandas as pd
```

```python
#reading the necessary file
df = pd.read_csv("/Users/souradipgoswami/Documents/DCUMaterials/CA682DataViz/Ranking.csv")
```

```python
df.head()
```

```python
#dropping the unnecessary columns
to_drop = ['rank','country_abrv','previous_points','rank_change','cur_year_avg','cur_year_avg_weighted','last_year_avg','last_year_avg_weighted','two_year_ago_avg','two_year_ago_weighted','three_year_ago_avg','three_year_ago_weighted','confederation']
```

```python
df.drop(to_drop, inplace= True, axis = 1)
```

```python
#splitting the rank date columns
df['rank_year']=[d.split('/')[2] for d in df.rank_date]
df['rank_month']=[d.split('/')[1] for d in df.rank_date]
```

```python
df.head()
```

```python
#converting the rank month to integer
df['rank_month1'] = df['rank_month'].map(int)
df.head()
```

```python
#extracting the month abbreviations
df.loc[df.rank_month1 == 1, 'month_abbr'] = 'January'
df.loc[df.rank_month1 == 2, 'month_abbr'] = 'February'
df.loc[df.rank_month1 == 3, 'month_abbr'] = 'March'
df.loc[df.rank_month1 == 4, 'month_abbr'] = 'April'
df.loc[df.rank_month1 == 5, 'month_abbr'] = 'May'
df.loc[df.rank_month1 == 6, 'month_abbr'] = 'June'
df.loc[df.rank_month1 == 7, 'month_abbr'] = 'July'
df.loc[df.rank_month1 == 8, 'month_abbr'] = 'August'
df.loc[df.rank_month1 == 9, 'month_abbr'] = 'September'
df.loc[df.rank_month1 == 10, 'month_abbr'] = 'October'
df.loc[df.rank_month1 == 11, 'month_abbr'] = 'November'
df.loc[df.rank_month1 == 12, 'month_abbr'] = 'December'
```

```python
df.head()
```

```python
#creating a column with month abbreviations and year
df['ranking_date'] = df['month_abbr'].map(str) + ',' + '20' + df['rank_year'].map(str)
```

```python
df.head()
```

```python
to_drop = ['rank_date','rank_year','rank_month','rank_month1','month_abbr']
```

```python
df.drop(to_drop, inplace = True, axis=1)
```

```python
df.head()
```

```python
#creating a new dataframe with each ranking date as columns and country as rows
new_df=pd.crosstab(df.country_full,df.ranking_date,values=df.total_points,aggfunc='sum')
```
