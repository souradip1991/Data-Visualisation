```python
#importing necessary package and loading the input data
import pandas as pd
import numpy as np
df = pd.read_csv("/Users/souradipgoswami/Documents/DCUMaterials/CA682DataViz/data.csv")
df.head()
```

```python
#dropping unnecessary columns
to_drop=['ID','Release Clause']
df.drop(to_drop,inplace=True,axis=1)
df.head()
```

Best Goalkeeper: 

```python
#assigning weightage: mi-most important vi-very important i-important
mi=5
vi=4
i=3
```

```python
#checking the top goalkeeper
df['Top_Goalkeeping_points']= (i*df.Crossing + i*df.BallControl + vi*df.Agility+ mi*df.Reactions + vi*df.Balance+ vi*df.Jumping +vi*df.Stamina +vi*df.Strength+ mi*df.GKPositioning + mi*df.GKDiving + mi*df.GKHandling + mi*df.GKKicking+ mi*df.GKReflexes)
new1= df.sort_values('Top_Goalkeeping_points',ascending=False)
to_drop=['Top_Goalkeeping_points']
new1.drop(to_drop,inplace=True,axis=1)
df.drop(to_drop,inplace=True,axis=1)
goalkeeper_df=new1.head(1)
goalkeeper_df.head()
```

4 Defenders- 2 Centre backs, One left back and One right back

```python
#top 2 center backs
df['cb_points'] = (i*df.Crossing + i*df.Finishing+ mi*df.HeadingAccuracy+ i*df.BallControl+ i*df.SprintSpeed + vi*df.Agility+ mi*df.Reactions + vi*df.Balance+ vi*df.Stamina + vi*df.Strength + mi*df.Interceptions + mi*df.Positioning + mi*df.Composure+ mi*df.Marking+ mi*df.StandingTackle+ mi*df.SlidingTackle)
new2 = df[(df['Position'] == 'LCB' )].sort_values('cb_points', ascending=False)
new3 = df[(df['Position'] == 'RCB')].sort_values('cb_points', ascending=False)
to_drop=['cb_points']
new2.drop(to_drop,inplace=True,axis=1)
new3.drop(to_drop,inplace=True,axis=1)
df.drop(to_drop,inplace=True,axis=1)
LCB_df=new2.head(1)
RCB_df=new3.head(1)
LCB_df.head()
```

```python
#top 2 side backs
df['sideback_points'] = (vi*df.Crossing+ i*df.ShortPassing+ i*df.LongPassing+ i*df.HeadingAccuracy+ i*df.BallControl+ i*df.SprintSpeed + vi*df.Agility+ mi*df.Reactions + vi*df.Balance+ vi*df.Stamina + vi*df.Strength + mi*df.Interceptions + mi*df.Positioning + vi*df.Acceleration+ mi*df.Composure+ mi*df.Marking+ mi*df.StandingTackle+ mi*df.SlidingTackle)
new4 = df[(df['Position'] == 'LB')].sort_values('sideback_points', ascending=False)
new5 = df[(df['Position'] == 'RB')].sort_values('sideback_points', ascending=False)
to_drop=['sideback_points']
new4.drop(to_drop,inplace=True,axis=1)
new5.drop(to_drop,inplace=True,axis=1)
df.drop(to_drop,inplace=True,axis=1)
LB_df=new4.head(1)
RB_df=new5.head(1)
LB_df.head()
```

3 Midfielders- 1 CDM, 1 LCM, 1 RCM

```python
# 4 best midfielders
df['cdm_points'] = (vi*df.Crossing + mi*df.ShortPassing+ mi*df.LongPassing+ mi*df.BallControl+ vi*df.Dribbling + vi*df.SprintSpeed + vi*df.Agility + vi*df.Balance+ vi*df.Stamina + vi*df.ShotPower+ vi*df.Strength + mi*df.Interceptions + mi*df.Vision + vi*df.Acceleration+ mi*df.Composure+ vi*df.LongShots )
df['LCM_points'] = (vi*df.Crossing + mi*df.ShortPassing+ mi*df.LongPassing+ mi*df.BallControl+ vi*df.Dribbling + vi*df.SprintSpeed + vi*df.Agility + vi*df.Balance+ vi*df.Stamina + vi*df.ShotPower+ vi*df.Strength  + mi*df.Vision + vi*df.Acceleration+ mi*df.Composure+ vi*df.FKAccuracy + vi*df.Curve + vi*df.Volleys )
df['RCM_points'] = (mi*df.Crossing + mi*df.ShortPassing+ mi*df.LongPassing+ mi*df.BallControl+ vi*df.Dribbling + vi*df.SprintSpeed + vi*df.Agility + vi*df.Balance+ vi*df.Stamina + vi*df.Strength + mi*df.Interceptions + mi*df.Vision + vi*df.Acceleration+ mi*df.Composure+ vi*df.LongShots )
new6 = df[(df['Position'] == 'CDM')].sort_values('cdm_points', ascending=False)
new7 = df[(df['Position'] == 'LCM')].sort_values('LCM_points', ascending=False)
new8 = df[(df['Position'] == 'RCM')].sort_values('RCM_points', ascending=False)
to_drop=['cdm_points','LCM_points','RCM_points']
new6.drop(to_drop,inplace=True,axis=1)
new7.drop(to_drop,inplace=True,axis=1)
new8.drop(to_drop,inplace=True,axis=1)
df.drop(to_drop,inplace=True,axis=1)
CDM_df=new6.head(1)
LCM_df=new7.head(1)
RCM_df=new8.head(1)
CDM_df.head()
```

3 strikers- 1 Right forward, 1 striker, 1 left forward

```python
#3 best strikers
df['lwrw_points'] = (mi*df.Finishing + mi*df.HeadingAccuracy+ vi*df.ShortPassing+ mi*df.Volleys+ mi*df.BallControl+ vi*df.Dribbling + vi*df.Curve+ vi*df.SprintSpeed + vi*df.Agility + vi*df.Balance+ vi*df.Stamina + mi*df.ShotPower+ vi*df.Strength  + vi*df.Acceleration+ mi*df.Composure + vi*df.Jumping )
df['st_points'] = (mi*df.Finishing + mi*df.HeadingAccuracy+ mi*df.Volleys+ mi*df.BallControl+ i*df.Dribbling + vi*df.Curve + vi*df.Agility + vi*df.Balance+ vi*df.Stamina + mi*df.ShotPower  + vi*df.Acceleration+ mi*df.Composure )
new9 = df[(df['Position'] == 'ST')].sort_values('st_points', ascending=False)
new10 = df[(df['Position'] == 'LF')].sort_values('lwrw_points', ascending=False)
new11 = df[(df['Position'] == 'RF')].sort_values('lwrw_points', ascending=False)
to_drop=['lwrw_points','st_points']
new9.drop(to_drop,inplace=True,axis=1)
new10.drop(to_drop,inplace=True,axis=1)
new11.drop(to_drop,inplace=True,axis=1)
df.drop(to_drop,inplace=True,axis=1)
ST_df=new9.head(1)
LF_df=new10.head(1)
RF_df=new11.head(1)
ST_df.head()
```

Merging all the new dataframes and creating the best eleven dataset

```python
final=pd.concat([goalkeeper_df,LB_df,LCB_df,RCB_df,RB_df,LCM_df,CDM_df,RCM_df,LF_df,ST_df,RF_df])
final.head(11)
```

```python
export_csv = final.to_csv ("/Users/souradipgoswami/Documents/DCUMaterials/CA682DataViz/Top11.csv", header=True)
```
